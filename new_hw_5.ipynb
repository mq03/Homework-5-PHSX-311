```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#constants:
#x=0.200 m
A_0 = x = 0.200 ##take out the x= part 
k = 2.50
m = 0.125
Q_ud = np.array([8.00, 2.00])
Q_cd = 0.5
Q_od = -0.5
```

```python
#calculate natural angular frequency
omega_0 = np.sqrt(k/m)
```

```python
#underdamped constant calculations,
gamma_ud = omega_0/Q_ud
omega_ud = np.sqrt(((omega_0)**2)-((gamma_ud/2)**2))
P_ud = (2*np.pi)/omega_ud #periods (the # here makes this a comment)
phi_ud = np.arctan2(-gamma_ud/2, omega_ud)
A_ud = (A_0)*(omega_0/omega_ud) # amplitude constant of the oscillator ('C_1' in class)
```

```python
#time settings for plot
tmin = 0
tmax = 2*np.min(P_ud)
nstep = 5000
```

```python
#set up time array
t = np.linspace(tmin, tmax, nstep)
```

```python
### oscillator eqns ###
#underdamped
x_2= A_ud[1]*(np.exp(-gamma_ud[1]*t/2))*np.cos(omega_ud[1]*t+phi_ud[1])
x_8= A_ud[0]*np.exp(-gamma_ud[0]*t/2)*np.cos(omega_ud[0]*t+phi_ud[0])
```

```python
np.power(2,2)
```

```python
#underdamped approx
x_2_app= A_0*np.exp(-gamma_ud[1]*t/2)*np.cos(omega_0*t)
x_8_app= A_0*np.exp(-gamma_ud[0]*t/2)*np.cos(omega_0*t)
```

```python
#plot x vs t UNDERDAMPED
plt.clf()
plt.plot(t, x_2, linestyle = '-', color = 'black', label = 'Q = ' + str(Q_ud[1]))
plt.plot(t, x_8, linestyle = '-', color = 'orange', label = 'Q = ' + str(Q_ud[0]))
plt.plot(t, x_2_app, color = 'black', label = 'Approximation', linestyle = ':')
plt.plot(t, x_8_app, color = 'orange', label = 'Approximation', linestyle = ':')
plt.axhline(y=0, linestyle = ':')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()
plt.show()
```

```python
#critically damped
gamma_cd = 2*omega_0
```

```python
#critically-damped
x= A_0*(1+(gamma_cd/2)*t)*np.exp(-gamma_cd*t/2)
```

```python
#plot x vs t CRITICALLY-DAMPED
plt.clf()
plt.plot(t, x, linestyle = '-', color = 'black', label = 'Q = ' + str(Q_cd))
#plt.plot(t, plt.plot, linestyle = '-', color = 'orange', label = 'Q = ' + str(Q_ud[1]))
#plt.plot(t, plt.plot, color = 'black', label = 'Approximation', linestyle = ':')
#plt.plot(t, plt.plot, color = 'orange', label = 'Approximation', linestyle = ':')
plt.axhline(y=0, linestyle = ':')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.legend()
plt.show()
```
